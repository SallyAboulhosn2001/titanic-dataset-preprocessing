```python
import pandas as pd 
titanic_data=pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
```

```python
#we'll take the median of the age column to fill the missing values
titanic_data['Age']=titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

```python
#we'll drop cabin column since it has many missing values
titanic_data=titanic_data.drop(columns=['Cabin'])
```

```python
titanic_data['Embarked']=titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0])
#mode() gives the most frequent values in a specific column, [0] tells to choose the first value which is the most frequent one
```

We convert to numbers most of the time if not always since models don't understand text

```python
titanic_data['Sex']=titanic_data['Sex'].map({'male':0,'female':1})
```

```python
titanic_data['Embarked']=titanic_data['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
titanic_data['FamilySize']=titanic_data['SibSp']+titanic_data['Parch']+1
#we added +1 for the passenger himself
```

**.loc[rows,columns], so below we selected rows =1 for familysize, and we set isAlone to 1 if the value in familysize row is 1.**

```python
titanic_data['isAlone']=0 #we created a column isAlone, 0 is false
titanic_data.loc[titanic_data['FamilySize']==1,'isAlone']=1 #now here, if familysize is equal to 1 this means that this passenger is alone, so set isAlone value to 1
```

```python
titanic_data=titanic_data.drop(columns=['PassengerId','Name','Ticket'])
```
